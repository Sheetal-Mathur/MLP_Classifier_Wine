```python
class MyNeuralNetwork:
  def __init__(self):
    self.w1=1
    self.w2=1
    self.bias=0

  def fit(self,x,y,epochs,loss_threshold):
    print(x['age'])
    print('---------------------------')
    print(x['afford'])
    self.w1,self.w2,self.bias= self.gradient_descent(x['age'],x['afford'],y,epochs,loss_threshold)
    print(f'final w1 : {self.w1} w2 : {self.w2} bias: {self.bias}')
  
  def myprediction(self,xtest):
    weightedsum= xtest['age']*self.w1 + xtest['afford']*self.w2+self.bias
    return sigmoid_function(weightedsum)
  
  def log_loss(self,ytrue,ypred):
    ep=1e-15
    ypred_new =[max(i,ep) for i in ypred]
    ypred_new =[min(i,1-ep)for i in ypred_new]
    ypred_new=np.array(ypred_new)
    return -np.mean(ytrue * np.log(ypred_new) + (1-ytrue) *(1-np.log(ypred_new)))


 
  def gradient_descent(self,age,afford,ytrue,epochs,loss_threshold):
    w1 = w2 = 1
    bias= 0
    rate=0.5
    n=len(age)
    for i in range(epochs):
      sumOfWeight = w1*age+w2*afford+bias
      ypredict=sigmoid_function(sumOfWeight)
      loss=(self.log_loss(np.array(ytrue['insurace']),ypredict))
      print(loss)
      
     
      w1d=1/n* np.dot(np.transpose(age),(ytrue['insurace']-ypredict))
      w2d=1/n* np.dot(np.transpose(afford),(ytrue['insurace']-ypredict))
      biasd=np.mean(ypredict-ytrue['insurace'])
      print(w1d,w2d,biasd)

      w1=w1-rate*w1d
      w2=w2-rate*w2d
      bias=bias-rate*biasd

      print(f'epoch : {i} w1 : {w1} w2 : {w2} bias: {bias} loss: {loss}')
      if loss<=loss_threshold:
        break
    return w1,w2,bias 


```

```python
mymodel= MyNeuralNetwork()
mymodel.fit(xtrain_scale,ytrain,epochs=1000,loss_threshold=.4416)
```
