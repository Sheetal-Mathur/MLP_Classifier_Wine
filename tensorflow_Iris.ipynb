```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
from sklearn.datasets import load_iris
dataset =load_iris()
```

```python
df_flower=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df_flower.head()
```

```python
df_flower['flower_label']=dataset['target']
df_flower.head()
```

```python
df_flower['flower_name']=df_flower['flower_label'].replace(to_replace=[0,1,2],value=dataset['target_names'])
```

```python
df_flower.head()
```

```python
from sklearn.model_selection import train_test_split
X=df_flower.iloc[:,0:4]
Y=df_flower.iloc[:,4:5]
xtrain,xtest,ytrain ,ytest  = train_test_split(X,Y,test_size=0.30,random_state=1)
```

```python
from keras.models import Sequential
from  keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(4,),activation="relu"))
```

```python
#from keras.backend import sparse_categorical_crossentropy
model.compile(
    optimizer="adam",
    loss='sparse_categorical_crossentropy',
    metrics=["accuracy"]

)
model.fit(xtrain,ytrain,epochs=200)
```
